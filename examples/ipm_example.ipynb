# Imports

```python
import matplotlib.pyplot as plt

from microlensing.IPM.ipm import IPM
```

# Creating a magnification map

```python
ipm = IPM(verbose=1, # controls level of output printed from the C++ library. 0 for nothing, 1 for some (e.g. progress bar), 2 for more, 3 for most
          kappa_tot=0.4, shear=0.4, kappa_star=0.2, # macromodel lensing parameterse
          half_length_y1=25, half_length_y2=25, # map size
          num_pixels_y1=5000, num_pixels_y2=5000
          )
```

```python
ipm.random_seed = 0 # setting to 0 will cause object to generate a random seed of its own when run
```

```python
ipm.run()
```

```python
ipm.num_stars
```

# Basic plotting

```python
# by default plots log mu
fig, ax = plt.subplots()
ipm.plot(ax)
plt.show()
```

```python
# can also plot astronomical magnitudes from the macro-model magnification
fig, ax = plt.subplots()
ipm.plot(ax, plot_magnitudes=True)
plt.show()
```

```python
# histograms can also be plotted, of log(mu) by default, or astronomical magnitudes
fig, ax = plt.subplots()
ipm.plot_hist(ax, plot_magnitudes=False, histtype='step')
plt.show()
```
